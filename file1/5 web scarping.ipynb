```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
driver.get('https://www.flipkart.com/') 
```

```python
search_box = driver.find_element(By.CLASS_NAME,"zDPmFV")
search_box.send_keys('sneakers')
```

```python
submit_button = driver.find_element(By.CSS_SELECTOR, "button[type='submit']")
submit_button.click()
```

```python

# Lists to store scraped data
brands = []
product_descriptions = []
prices = []

# Function to extract data from the current page
def extract_data():
    brand_elements = driver.find_elements(By.CLASS_NAME, "syl9yP")  
    product_description_elements = driver.find_elements(By.CLASS_NAME, "WKTcLC") 
    price_elements = driver.find_elements(By.CLASS_NAME, "Nx9bqj") 
    brands.extend([brand.text for brand in brand_elements])
    product_descriptions.extend([desc.text for desc in product_description_elements])
    prices.extend([price.text for price in price_elements])

# Extract from the first page
extract_data()

# Pagination to load more sneakers (up to 100 brands)
while len(brands) < 100:
    try:
        next_button = driver.find_elements(By.CLASS_NAME, "_9QVEpD")  # Class for the "Next" button
        
        if next_button:
            next_button[0].click()  # Click the first next button found
            time.sleep(5)  # Wait for the new sneakers to load
            extract_data()  # Extract data from the new page
        else:
            print("No more pages available.")
            break
    except Exception as e:
        print(f"Error while clicking next: {e}")
        break

max_length = max(len(brands), len(product_descriptions), len(prices))

# Extend lists to the maximum length
brands.extend([''] * (max_length - len(brands)))
product_descriptions.extend([''] * (max_length - len(product_descriptions)))
prices.extend([''] * (max_length - len(prices)))


# Create a DataFrame from the collected data
data = {
    'Brand': brands[:100],  # Limit to first 100
    'Product Description': product_descriptions[:100],  # Limit to first 100
    'Price': prices[:100]  # Limit to first 100
}

df = pd.DataFrame(data)

# Display the DataFrame
print(df)


```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# Lists to store scraped data
quotes = []
authors = []
types = []

# Function to extract data from the current page
def extract_data():
    quote_elements = driver.find_elements(By.CLASS_NAME, 'title')
    author_elements = driver.find_elements(By.CLASS_NAME, 'author')
    type_elements = driver.find_elements(By.CLASS_NAME, 'tags')

    quotes.extend([quote.text.strip() for quote in quote_elements])
    authors.extend([author.text.strip() for author in author_elements])
    types.extend([', '.join(tag.text.strip() for tag in type_element.find_elements(By.TAG_NAME, 'a')) for type_element in type_elements])

# Setup the Selenium WebDriver
driver = webdriver.Chrome()
 

# Start with the first page
page_number = 1
driver.get(f"https://www.azquotes.com/top_quotes.html?p={page_number}")
time.sleep(2)  # Wait for the page to load
extract_data()

# Pagination to load more quotes (up to 1000 quotes)
while len(quotes) < 1000:
    try:
        # Navigate to the next page
        page_number += 1
        driver.get(f"https://www.azquotes.com/top_quotes.html?p={page_number}")
        time.sleep(2)  # Wait for the new page to load
        extract_data()  # Extract data from the new page
    except Exception as e:
        print(f"Error while fetching page {page_number}: {e}")
        break

# Ensure all lists are of the same length
max_length = max(len(quotes), len(authors), len(types))
quotes += [None] * (max_length - len(quotes))
authors += [None] * (max_length - len(authors))
types += [None] * (max_length - len(types))

# Create a DataFrame
df = pd.DataFrame({
    'Quote': quotes[:1000],
    'Author': authors[:1000],
    'Type': types[:1000]
})

# Drop rows where 'Quote' or 'Author' is None
df = df.dropna(subset=['Quote', 'Author'])

# Print the cleaned DataFrame
print(df)


```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
```

```python
driver = webdriver.Chrome()
driver.get('https://www.jagranjosh.com/general-knowledge/list-ofall-prime-ministers-of-india-1473165149-1') 
```

```python
# Lists to store scraped data
names = []
born_dead = []
terms = []
remarks = []

# Find the div containing the table data
table_data = driver.find_element(By.CLASS_NAME, 'TableData')

# Iterate through each row in the div
for row in table_data.find_elements(By.TAG_NAME, 'tr')[1:]:  # Skip the header row
    cols = row.find_elements(By.TAG_NAME, 'td')  # Get all columns in the row
    if len(cols) >= 4:  # Ensure there are enough columns
        names.append(cols[0].text.strip())
        born_dead.append(cols[1].text.strip())
        terms.append(cols[2].text.strip())
        remarks.append(cols[3].text.strip())

# Create a DataFrame from the scraped data
df = pd.DataFrame({
    'Name': names,
    'Born-Dead': born_dead,
    'Term of Office': terms,
    'Remarks': remarks
})

# Display the DataFrame
print(df)
```

```python
driver = webdriver.Chrome()
driver.get("https://www.motor1.com/")
```

```python
search_bar = driver.find_element(By.CLASS_NAME, "m1-search-panel-input")  # Use the specified class for the search bar
search_bar.send_keys("50 most expensive cars")  # Type the search query

# Click the search button
search_button = driver.find_element(By.CLASS_NAME, "m1-search-panel-button")  # Use the specified class for the button
search_button.click()  
```

```python
car_link = driver.find_element(By.XPATH, "//a[@href='/features/308149/most-expensive-new-cars-ever/']")
car_link.click()
```

```python
# Find the container by ID
car_listings = driver.find_element(By.ID, "page_index_articles_detail")

# Initialize lists for car names and prices
car_names = []
car_prices = []

# Extract car names and prices directly from the car_listings
try:
    # Extract car names
    name_elements = car_listings.find_elements(By.CLASS_NAME, "subheader")
    price_elements = car_listings.find_elements(By.TAG_NAME, "strong")
    
    for name, price in zip(name_elements, price_elements):
        try:
            car_names.append(name.text.strip())
            car_prices.append(price.text.strip())
        except Exception as e:
            print(f"Error extracting data: {e}")

except Exception as e:
    print(f"Error finding car listings: {e}")

# Create a DataFrame from the scraped data
df = pd.DataFrame({
    'Car Name': car_names,
    'Price': car_prices
})

# Print the DataFrame
print(df)
```
